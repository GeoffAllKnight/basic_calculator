```python
import math
from IPython.display import clear_output

initial_num = 0
second_num = 0
operation = 0
calc_on = True
continuous = True
first_run = True

def add_num(initial_num,second_num):
    return initial_num + second_num

def minus_num(initial_num,second_num):
    return initial_num - second_num

def multiply_num(initial_num,second_num):
    return initial_num * second_num

def divide_num(initial_num,second_num):
    return initial_num / second_num

def power_num(initial_num,second_num):
    return initial_num ** second_num

def sqroot_num(initial_num):
    return initial_num**0.5

def percent_num(initial_num):
    return initial_num / 100

def pi_num():
    return math.pi

def pos_2_neg(final_num):
    if final_num > 0:
        return -(final_num)
    elif final_num < 0:
        return abs(final_num)

def again():
    global calc_on
    global first_run
    
    again_or_not = str(input("Would you like to run a new equation? (Y or N): "))

    while again_or_not.upper() != "Y" and again_or_not.upper() != "N":
        again_or_not = str(input("Wrong input y'all (Y or N): "))
    if again_or_not == "N":
        calc_on = False
        clear_output()
        print("Thanks for using my calculator!")
    else:
        first_run = True
        clear_output()

#Beginning of calculator function

print("Welcome to this basic AF calculator!")
while calc_on:
    
    while first_run:
        print("List of Available Operations:\n1. Addition\n2. Subtraction\n3. Multiplication\n4. Division\n5. Power\n6. Square Root\n7. Percentage\n8. Pi")
        
        while operation not in range(1,9):
            try:
                print("Type the number for the operation you'd like to do")
                operation = int(input("What type of operation would you like to do? "))
            except:
                print("That is an invalid-  operation. Please enter a valid number for operation.")
            if operation not in range(1,9):
                print("Not a valid operation number!")
            else:
                break
                
        #Start of all the operations selections
        while first_run:
            if operation in [1,2,3,4,5]:
                while True:
                    try:
                        initial_num = int(input("Enter your first number: "))
                        second_num = int(input("Enter your second number: "))
                        first_run = False
                        break
                    except:
                        print("That is an invalid number")
            elif operation in [6,7]:
                while True:
                    try:
                        initial_num = int(input("Enter your number: "))
                        first_run = False
                        break
                    except:
                        print("That is not a valid number!")
    while continuous:     
        if operation == 1:
            final_num = add_num(initial_num,second_num)
            print(f"{initial_num} + {second_num} = {final_num}")
        elif operation == 2:
            final_num = minus_num(initial_num,second_num)
            print(f"{initial_num} - {second_num} = {final_num}")
        elif operation == 3:
            final_num = multiply_num(initial_num,second_num)
            print(f"{initial_num} x {second_num} = {final_num}")
        elif operation == 4:
            final_num = divide_num(initial_num,second_num)
            print(f"{initial_num} / {second_num} = {final_num}")
        elif operation == 5:
            final_num = power_num(initial_num,second_num)
            print(f"{initial_num} to the power of {second_num} = {final_num}")
        elif operation == 6:
            final_num = sqroot_num(initial_num)
            print(f"The square root of {initial_num} is: {final_num}")
        elif operation == 7:
            final_num = percent_num(initial_num)
            print(f"The percentage conversion of your number is {final_num}%") 
        elif operation == 8: 
            final_num = pi_num()
            print(f"Pi is {final_num}")

        operation = 0
        while operation not in range(1,11):
            print(f"Your current calculated number: {final_num}")
            print("Would you like to further perform of the following functions?")
            print("List of Available Operations:\n1. Addition\n2. Subtraction\n3. Multiplication\n4. Division\n5. Power\n6. Square Root\n7. Percentage\n8. Pi\n9. Convert +/-\n0. No")
            try:
                operation = int(input("Enter an operation number here: "))
                clear_output()
            except:
                print("That is not a valid operation!")
            if operation == 9:
                final_num = pos_2_neg(final_num)
                print(f"The result is {final_num}")
            if operation == 0:
                continuous = False
                break
            initial_num = final_num
    again()
    
```
